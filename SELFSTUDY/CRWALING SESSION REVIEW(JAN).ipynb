```python
# module importing
import requests                  # requests 데이터 받아오는 역할하는 라이브러리
from bs4 import BeautifulSoup    # BeautifulSoup 함수
```

```python
# 데이터를 얻고 싶은 웹사이트 주소 url 변수에 저장
url = "http://naver.com"    
```

```python
# 특정 주소에 데이터 요청하고 변수에 저장
response = requests.get(url)
# 따로 저장하지 않았다면 바로 주소입력가능
# 변수 = requests.get("웹주소")
# response = requests.get("http://naver.com")
print(response)
```

```python
# 해당 웹사이트의 html 코드 보기
print(response.text)
```

```python
# HTML코드를 변수에 저장
html = response.text       
```

```python
# BeautifulSoup 메소드 활용해 해당 메소드가 다룰 수 있는 타입으로 바꿔줌
naver = BeautifulSoup(html, "html.parser")
```

```python
#  HTML코드의 범위축소 - find / find_all
# 원하는 속성을 가진 특정 태그를 검색시
# find find_all 메소드
# 1줄만 가져올 때 
naver.find("span", class_= 'ah_k')
```

```python
# BeautifulSoup로타입바꿔서저장한변수.find("태그", {'class' : '속성'}) 
# 지정 속성과 태그에 해당하는 데이터 다 가져올 때 
# 태그와 속성도 데이터와 함께 리스트 형태로 저장됨
# <span class="ah_k">안희정</span>

naver_search = naver.find_all("span", class_= 'ah_k')
# 중괄호로 묶어주지 않아도 괜찮음
# ("태그", {'class' : '속성'})
# ("태그", class_= '속성')
```

```python
print(naver_search)
```

```python
# text만 뽑아내기
# python에서 인덱싱 번호 0부터시작
naver_search[0].text
```

```python
len(naver_search)        # 데이터 길이 len()
```

```python
range(len(naver_search)) # 인덱스 범위 range(len())
```

```python
for i in naver_search:
    print(i.text)        # 두번씩 나온다.
```

```python
# 한번 씩만 나오게 20위까지 뽑아내고싶다.
for i in range(len(naver_search)):
    if i < 20:
        print(naver_search[i].text)
```

```python
# 리스트 선언 후 집어넣기
naver_search_list = []
for i in range(len(naver_search)):
    if i <20 :
        naver_search_list.append(naver_search[i].text)    # apppend 메소드
```

```python
# [표현식 for 항목1 in 반복가능객체1 if 조건문1
#        for 항목2 in 반복가능객체2 if 조건문2
#        ...
#        for 항목n in 반복가능객체n if 조건문n]
```

```python
# naver_search_list2 = [i.text for i in range(len(naver_search))] 아님 표현식 확인
naver_search_list2 = [i.text for i in naver_search ]
print(naver_search_list2)    # 두번씩 나온다
```

```python
naver_search_list3 = naver_search.text    # 원소하나당 적용되는 메소드 text
naver_search_list3 
```
