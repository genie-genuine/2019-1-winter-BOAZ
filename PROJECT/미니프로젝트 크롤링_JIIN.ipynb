```python
# 1. module importing
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
df1=pd.read_html('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx', flavor='html5lib') 
```

```python
df1=df1[0]
```

```python
df1
```

```python
df1.columns
```

```python
# 2. driver경로설정
driverpath =  'C:/Users/USER/Downloads/chromedriver_win32/chromedriver.exe'
dr = webdriver.Chrome(driverpath)
```

```python
# url 주소 설정
dr.get('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx' )
dr.find_element_by_id('cphContents_cphContents_cphContents_ucPager_btnNo2').click()
```

```python
list = dr.find_elements_by_class_name('record_result')
for data in list:
    df2_table = data.find_element_by_tag_name('table').text
df2_table
```

```python
df2_rows = np.array(str(df2_table.split('\n')[1]).split(' ')).reshape(1,16)
df2_rows.shape
```

```python
columns = df1.columns
df2 = pd.DataFrame(columns=columns)
for i in range(1,31):
    df2_rows = np.array(str(df2_table.split('\n')[i]).split(' ')).reshape(1,16)
    df2 = df2.append(pd.DataFrame(df2_rows,columns=columns), ignore_index=True)
df2
```

```python
dr.get('https://www.koreabaseball.com/Record/Player/HitterBasic/Basic1.aspx' )
dr.find_element_by_id('cphContents_cphContents_cphContents_ucPager_btnNo3').click()
```

```python
dr.find_element_by_id('cphContents_cphContents_cphContents_ucPager_btnNo3').click()
list = dr.find_elements_by_class_name('record_result')
for data in list:
    df3_table = data.find_element_by_tag_name('table').text
df3_table
```

```python
columns = df1.columns
df3 = pd.DataFrame(columns=columns)
for i in range(1,3):
    df3_rows = np.array(str(df3_table.split('\n')[i]).split(' ')).reshape(1,16)
    df3 = df3.append(pd.DataFrame(df3_rows,columns=columns), ignore_index=True)
df3
```

```python
kbo_record=pd.concat([df1,df2,df3])
kbo_record.to_excel("KBO_record.xlsx", encoding='utf-8', index=False)
```
