```python
# 1. module importing
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
defense=pd.read_html('https://www.koreabaseball.com/Record/Player/Defense/Basic.aspx')
defense=defense[0]
defense.columns
```

```python
position1 = defense[['순위','선수명','팀명', 'POS']]
position1
```

```python
driverpath =  'C:/Users/USER/Downloads/chromedriver_win32/chromedriver.exe'
dr = webdriver.Chrome(driverpath)
dr.get('https://www.koreabaseball.com/Record/Player/Defense/Basic.aspx' )
```

```python
dr.find_element_by_id('cphContents_cphContents_cphContents_ucPager_btnNo2').click()
```

```python
list = dr.find_elements_by_class_name('record_result')
for data in list:
    defense2_table = data.find_element_by_tag_name('table').text
defense2_table
```

```python
columns = position1.columns
position = pd.DataFrame(columns=columns)
columns
```

```python
position
```

```python
for i in range(1,31):
    position_rows = np.array(defense2_table.split('\n')[i].split()[0:4]).reshape(1,4)
    position = position.append(pd.DataFrame(position_rows, columns=columns))
position
```
