```python
import requests                 # 데이터 받아오기
from bs4 import BeautifulSoup   # 데이터에서 값을 뽑아내는 역할
import pandas as pd
import time
```

```python
url = "http://zum.com"          # 데이터를 얻고 싶은 주소
response = requests.get(url)    # 특정 주소에 GET 요청
html = response.text            # HTML 소스 가져오기
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, "html.parser") 
# 첫 인자는 html 소스코드, 두 번째 인자는 어떤 parser 이용할지 명시
```

```python
print(soup.prettify())            #
```

```python
# a = soup.find("ul", {'class' : ah_l})
#                태그,  속성
jiin = soup.find("span", {'class':'keyword d_keyword'} )
print(jiin)
```

```python
# soup.find_all() 중요!
jiin1 = soup.find_all('span', {'class' : 'keyword d_keyword'})
print(jiin1)
```
