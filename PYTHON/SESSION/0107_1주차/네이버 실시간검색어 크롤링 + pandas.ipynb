## bs4

```python
#url = "https://www.naver.com" #데이터를 얻고 싶은 주소
url = "https://www.daum.net/" #데이터를 얻고 싶은 주소
response = requests.get(url) #특정 주소에 GET 요청
html = response.text #HTML소스 가져오기
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, "html.parser") 
#첫 인자는 html소스코드, 두 번째 인자는 어떤 parser를 이용할지 명시.
```

```python
print(soup.prettify())
```

```python
#hyunseok = soup.find("ul", {'class' : 'ah_l'}) #첫번째 인자 : 태그, 두번째 인자 : 속성
#구조를 이해해야한다.

hyunseok = soup.find("span", {'class' : 'txt_issue'})

print(hyunseok)
#우리가 뽑아내야 하는 것 : 순위, 검색어
```

```python
hyunseok1 = soup.find_all('span', {'class' : 'txt_issue'})
print(hyunseok1)
```

```python
hyunseok1[0].text
```

```python
len(hyunseok1)
```

```python
for i in hyunseok1 :
    print(i.text)    # for문 이용해서 정제할 데이터 확인해보기
```

```python
hyunseok1 = soup.find_all('span', {'class' : 'ah_k'})
hyunseok2 = soup.find_all('span', {'class' : 'ah_r'})
```

```python
hyunseok1
```

```python
hyunseok2
```

```python
search_word = []
rank = []
for i,j in zip(hyunseok1, hyunseok2) :
    search_word.append(i.text)
    rank.append(j.text)

#zip함수 : 동일한 개수로 이루어진 자료형을 묶어 주는 역할을 하는 함수
```

```python
date = time.strftime("%y%m%d",time.localtime())
#첫번째 인자 : 출력양식(format) Shift + tab 참고
#두번째 인자 : 시간
```

```python
search_word
```

## pandas

```python
#생성
df = pd.DataFrame({'search_word' : search_word, 'rank' : rank, 'date' : date}) #feature명 : data 를 dict형태로 나열.
```
